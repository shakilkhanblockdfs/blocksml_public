```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # Generate 100 random data points
y = 4 + 3 * X + np.random.randn(100, 1)  # Linear relationship with some noise

```

```python
# Visualize the data
plt.scatter(X, y, alpha=0.6)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter Plot of Synthetic Data")
plt.show()

```

```python
# Create a LinearRegression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Retrieve the model parameters
slope = model.coef_[0][0]
intercept = model.intercept_[0]

# Print the model parameters
print("Slope (Coefficient):", slope)
print("Intercept:", intercept)

# Make predictions using the trained model
X_new = np.array([[0], [2]])  # New data points for prediction
y_pred = model.predict(X_new)

print("------------")
print(X_new)
print(y_pred)
```

```python
# Visualize the linear regression line
plt.scatter(X, y, alpha=0.6, label="Original Data")
plt.plot(X_new, y_pred, color="red", linewidth=2, label="Linear Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```
