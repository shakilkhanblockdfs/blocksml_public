```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df
```

```python
df.shape
```

```python
# Using get dummies
df_dummies = pd.get_dummies(df['town'], drop_first=True) # Will create n-1 hot encong only
df_dummies
```

```python
df_dummies.shape
```

```python
df_new = pd.concat([df,df_dummies], axis='columns')
df_new
```

```python
df_new.shape
```

```python
df_new = df_new.drop(['town'], axis=1)
df_new
```

```python
X = df_new.drop('price', axis=1)
X
```

```python
y= df_new['price']
y
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
X
```

```python
model.predict([[3600,0,1]])
```

```python
model.score(X,y)
```
