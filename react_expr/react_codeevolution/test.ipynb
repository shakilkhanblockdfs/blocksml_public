```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f_exp(x):
    return 2**x - 1

def f_sqrt(x):
    return np.sqrt(x)

# Define the x points for both functions
x_points_exp = np.array([-8, -7, -6, -5, -4, -3, -2, -1, 0, 1,2,3,4], dtype=float)
x_points_sqrt = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=float)

# Calculate the y points for both functions
y_points_exp = f_exp(x_points_exp)
y_points_sqrt = f_sqrt(x_points_sqrt)

# Create the combined plot
plt.figure(figsize=(10, 6))
plt.plot(x_points_exp, y_points_exp, marker='o', linestyle='-', color='b', label='f(x) = 2^x - 1')
plt.plot(x_points_sqrt, y_points_sqrt, marker='o', linestyle='-', color='g', label='f(x) = sqrt(x)')
plt.title('Graphs of f(x) = 2^x - 1 and f(x) = sqrt(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
