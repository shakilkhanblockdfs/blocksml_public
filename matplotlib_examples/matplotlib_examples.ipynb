```python
# Violin Plot       # it shows the data which is concentrated at some location

df = pd.read_csv("Automobile.csv")
sns.violinplot(data=df, x="horsepower")
sns.violinplot(data=df, y="horsepower")                               # Make a vertical graph
sns.violinplot(data=df, x="engine_size", y="fuel_type")                 # Make a vertical graph
sns.violinplot(data=df, x="body_style", y="engine_size", palette="bright")                 # Make a vertical graph
sns.violinplot(data=df, x="body_style", y="engine_size", palette="colorblind")                 # Make a vertical graph
plt.show()
```

```python
# Strip Plot

df = pd.read_csv("Automobile.csv")
sns.stripplot(data=df, x="engine_size")
plt.show()

plt.figure(figsize=(15,7))
sns.violinplot(data=df, x="body_style", y="engine_size")      # Violin plot for the same set of data. compare violin plot with strip plot
sns.stripplot(data=df, x="body_style", y="engine_size")         # Strip plot for the same segmented data
sns.stripplot(data=df, x="body_style", y="engine_size", jitter = True)         # Strip plot for the same segmented data
sns.stripplot(data=df, x="body_style", y="engine_size", jitter = 0.25)         # Strip plot for the same segmented data
sns.stripplot(data=df, x="fuel_type", y="engine_size", hue="number_of_doors", jitter = 0.25)         # Strip plot for the same segmented data
plt.show()
```

```python
# Swarm plot

df = pd.read_csv("Automobile.csv")
# sns.swarmplot(data=df, x="number_of_doors", y="price")
# sns.swarmplot(data=df, x="fuel_type", y = "price", hue="number_of_doors")
sns.swarmplot(data=df, x="fuel_type", y = "price", hue="number_of_doors", dodge=True)
plt.show()
```

```python
# Cat plot

'''
Seaborn's catplot() function is a versatile function that can create several different types of categorical plots. 
By default, if no kind is specified, catplot() will create a strip plot. However, by using the kind parameter, 
you can create different types of plots. The options for kind include:
"strip" (the default)
"swarm"
"box"
"violin"
"boxen"
"point"
"bar"
'''

df = pd.read_csv("Automobile.csv")
sns.catplot(data=df, x="fuel_type", y="horsepower")
sns.catplot(data=df, x="body_style", y="horsepower", hue="fuel_type", kind="point")
sns.catplot(data=df, x="fuel_type", y="horsepower", hue="number_of_doors", col="drive_wheels", kind="bar", palette="pastel")
sns.catplot(data=df, x="fuel_type", y="engine_size", hue="body_style", col="number_of_doors", kind="box", palette="bright")
sns.catplot(data=df, x="number_of_doors", y="price", hue="body_style", col="fuel_type", kind="swarm", palette="muted")
plt.show()
```

```python
# Plotly details
import plotly.express as px

df = pd.read_csv("Automobile.csv")
his = px.histogram(df, x="price")        # Histogram using plotly
his.show()

# bar plot using plotly
bar = px.bar(df, x="peak_rpm", y="horsepower")
bar.show()

# scatter plot using plotly
scat = px.scatter(df, x="price", y="engine_size")
scat.show()

fig = px.box(df, x="fuel_type", y="horsepower", points="all")
fig = px.box(df, x="fuel_type", y="horsepower")
fig.show()

# 3D scatter plot
fig_3d = px.scatter_3d(df, x = "fuel_type", y="horsepower", z ="price", color="horsepower")
fig_3d.show()
fig_3d.write_html("scatter_3d.html")       # Save the plot in html file
```
