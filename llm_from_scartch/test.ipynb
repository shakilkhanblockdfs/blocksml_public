```python
sentence =  "How how are you doing, let see if this makes sence. I will encoding this"
```

```python
sentence_List = sentence.split()
```

```python
sentence_List
```

```python
vocab = {word:idx for idx,word in enumerate(sentence_List)}
```

```python
vocab
```

```python
class tokenizer:
    def __init__(self, vocab):
        self.vocab = vocab
        self.idx_vocab = { idx:word for word, idx in self.vocab.items()}
    def encode(self, sentence):
        return [self.vocab[word] for word in sentence.split()]
    def decode(self, token_ids):
        result = []
        for id in token_ids:
            result.append(self.idx_vocab[id])
        return result
```

```python
t = tokenizer(vocab)
print(t.encode("How are you"))
```

```python
print(t.decode([1,4,6]))
```
