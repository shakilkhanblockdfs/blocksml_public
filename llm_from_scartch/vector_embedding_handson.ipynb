```python
import gensim.downloader as api
import numpy as np
model = api.load("word2vec-google-news-300")
```

```python
word_vectors = model
print(len(word_vectors['computer']))
print((word_vectors['computer'].shape))
print((word_vectors['computer']))
```

```python
word_vectors.most_similar(positive=['king','woman'], negative=['man'])
```

```python
print(word_vectors.similarity('woman','man'))
print(word_vectors.similarity('king','queen'))
print(word_vectors.similarity('boy','girl'))
print(word_vectors.similarity('monkey','banana'))
print(word_vectors.similarity('chimpanzee','banana'))
print(word_vectors.similarity('lion','banana'))
```

```python
print(word_vectors.most_similar("tower", topn=5))
```

```python
vector_difference = word_vectors['king'] - word_vectors['man']
magnitude = np.linalg.norm(vector_difference)
print("Magnitude of the difference vector between 'king' and 'man':", magnitude)
```

```python
vector_difference = word_vectors['semiconductor'] - word_vectors['earthworm']
magnitude = np.linalg.norm(vector_difference)
print("Magnitude of the difference vector between 'king' and 'man':", magnitude)
```

```python
import torch
```

```python
input_ids = torch.tensor([2,3,5,1])
print(input_ids)                        
```

```python
vocan_size = 6
output_dim = 3
torch.manual_seed(123)
embedding_layer = torch.nn.Embedding(vocan_size, output_dim)
print(embedding_layer.weight)
```

```python
print(embedding_layer(torch.tensor([3])))
```

```python
print(embedding_layer(input_ids))
```
