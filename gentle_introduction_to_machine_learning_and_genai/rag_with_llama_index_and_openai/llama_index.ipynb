```python
# from llama_index.llms import OpenAI
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from IPython.display import Markdown, display
```

```python
documents = SimpleDirectoryReader("data").load_data()    
```

```python
documents
```

```python
index = VectorStoreIndex.from_documents(documents)
# print(type(index))
```

```python
index.storage_context.persist()       # Makes the index persists and it will create a sorage folder to store those
                                      # This is not necessary but is important if we want to persist the index which are created    
```

```python
index
```

```python
query_engine = index.as_query_engine()
```

```python
dir(index)
```
